Inspired by a [challenge](https://github.com/ajstarks/dubois-data-portraits/tree/master/challenge) to recreate some of Du Bois's classic visualisations.

Comparing 2 time series with a vertical orientation and a reversed axis. A clean red, white and black look.

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('db.mplstyle')
```

```python
# The Data
df = pd.read_csv('data/Dubois1.csv')

y = df['Year']
ts1, ts2 = 'Black','White'
x1, x2 = df[ts1], df[ts2]

xlim = [100,0]
ylim = [1790, 1890]
xticks = list(range(100,-5,-5))
yticks = list(range(1790,1900,10))

title = 'COMPARATIVE INCREASE OF BLACK AND WHITE\nPOPULATION OF GEORGIA.'
xlabel = 'PERCENTS'
```

```python
fig,ax = plt.subplots(figsize=(6,8))

ax.set_xlim(xlim)
ax.set_ylim(ylim)
ax.set_xticks(xticks)
ax.set_yticks(yticks)
ax.set_xticklabels(xticks, fontsize=8, alpha=.6)
ax.set_yticklabels(yticks, fontsize=10, alpha=.6)

ax.grid(axis='both', ls='-', lw=.5, c='r')

plt.plot(x1,y,ls='-', color='k', label=ts1)
plt.plot(x2,y,ls='--', color='k', label=ts2)
ax.legend(loc='upper center', fontsize=8, bbox_to_anchor=(0.5, -0.1), 
          ncol=2, columnspacing = 25,frameon=False)
ax.text(.5,-.1,xlabel, ha = 'center', transform=ax.transAxes, fontsize=8, alpha=.6)

fig.suptitle(title, fontsize=14, ha='center')

plt.savefig("vis/D1_dubois1.png")
#plt.savefig("vis/D1_dubois1.pdf")
plt.show()
```
